```python
# Step 2: Import the required classes and functions

from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
```

```python
# Step 3: Load a free text generation model
# We use distilgpt2 because it is smaller and easier to run in Colab

text_generator = pipeline(
    task="text-generation",
    model="distilgpt2",
    max_new_tokens=50
)

# Convert the Hugging Face pipeline into a LangChain LLM
llm = HuggingFacePipeline(pipeline=text_generator)
```

```python
# Step 4: Basic LLM call
# This is the simplest example of using LangChain with a language model

basic_output = llm.invoke("Explain Artificial Intelligence in simple terms.")

print("Basic LLM Output:")
print(basic_output)
```

```python
# Step 5: Create a reusable PromptTemplate
# PromptTemplate helps us avoid hardcoding prompts again and again

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

# Format the prompt by inserting a topic
formatted_prompt = prompt_template.format(topic="Machine Learning")

print("Formatted Prompt:")
print(formatted_prompt)
```

```python
# Step 6: Send the formatted prompt to the model

prompt_output = llm.invoke(formatted_prompt)

print("PromptTemplate Output:")
print(prompt_output)
```

```python
# Step 7: Create a prompt template with multiple input variables

multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

# Test the template with sample values
multi_prompt_1 = multi_input_template.format(
    topic="AI",
    audience="beginners",
    tone="friendly"
)

multi_prompt_2 = multi_input_template.format(
    topic="Python",
    audience="students",
    tone="fun"
)

print("Multi-Input Prompt 1:")
print(multi_prompt_1)

print("\nMulti-Input Prompt 2:")
print(multi_prompt_2)
```

```python
# Step 8: Run one multi-input prompt through the model

multi_output = llm.invoke(multi_prompt_1)

print("Multi-Input Prompt Output:")
print(multi_output)
```

```python
# Step 9: Create different prompt styles for the same topic

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}."
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a short story."
)

# Common topic
topic_name = "Neural Networks"

# Format all three prompts
teaching_prompt = teaching_template.format(topic=topic_name)
interview_prompt = interview_template.format(topic=topic_name)
story_prompt = story_template.format(topic=topic_name)

print("Teaching Prompt:")
print(teaching_prompt)

print("\nInterview Prompt:")
print(interview_prompt)

print("\nStory Prompt:")
print(story_prompt)
```

```python
# Step 10: Run the three prompt variations through the model

teaching_output = llm.invoke(teaching_prompt)
interview_output = llm.invoke(interview_prompt)
story_output = llm.invoke(story_prompt)

print("Teaching Output:")
print(teaching_output)

print("\nInterview Output:")
print(interview_output)

print("\nStory Output:")
print(story_output)
```

```python
# Step 11: Build a simple chain manually
# Here, we combine two steps:
# 1. create a prompt
# 2. send it to the model

def simple_chain(topic):
    """
    This function creates a prompt using PromptTemplate
    and then sends that prompt to the language model.
    """
    prompt = prompt_template.format(topic=topic)
    response = llm.invoke(prompt)
    return response

# Test the simple chain
chain_result = simple_chain("Deep Learning")

print("Simple Chain Output:")
print(chain_result)
```

```python
# Step 12: Create an input validation function
# This helps control user input before building prompts

def validate_inputs(audience, tone):
    """
    This function checks whether the audience and tone are valid.
    If not, it assigns default values.
    """

    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        print(f"Invalid audience '{audience}' found. Using default value 'beginner'.")
        audience = "beginner"

    if tone not in valid_tones:
        print(f"Invalid tone '{tone}' found. Using default value 'formal'.")
        tone = "formal"

    return audience, tone

# Test validation
validated_audience, validated_tone = validate_inputs("kids", "friendly")

print("Validated Audience:", validated_audience)
print("Validated Tone:", validated_tone)
```

```python
# Step 13: Create a dynamic prompt generator
# This function combines validation + template generation

def generate_prompt(topic, audience, tone, style):
    """
    This function validates the inputs
    and creates a structured prompt.
    """

    # Validate audience and tone
    audience, tone = validate_inputs(audience, tone)

    # Check style
    valid_styles = ["teaching", "interview", "storytelling"]
    if style not in valid_styles:
        print(f"Invalid style '{style}' found. Using default value 'teaching'.")
        style = "teaching"

    # Convert audience into natural words
    if audience == "beginner":
        audience_text = "beginners"
    elif audience == "intermediate":
        audience_text = "intermediate learners"
    else:
        audience_text = "experts"

    # Create the final template
    generator_template = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style."
    )

    # Format the final prompt
    final_prompt = generator_template.format(
        topic=topic,
        audience=audience_text,
        tone=tone,
        style=style
    )

    return final_prompt

# Test the function
generated_prompt = generate_prompt(
    topic="Natural Language Processing",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print("Generated Prompt:")
print(generated_prompt)
```

```python
# Step 14: Send the generated prompt to the model

generated_output = llm.invoke(generated_prompt)

print("Generated Prompt Output:")
print(generated_output)
```

```python
# Step 15: Build a very simple memory system
# We store previous user inputs in a Python list

conversation_memory = []

def memory_chat(user_input):
    """
    This function stores the conversation history
    and sends the full history to the model.
    """
    conversation_memory.append(user_input)

    # Join all previous user messages into one text block
    memory_text = "Conversation so far: " + " | ".join(conversation_memory)

    response = llm.invoke(memory_text)
    return response

# Test memory system
memory_output_1 = memory_chat("Hello")
memory_output_2 = memory_chat("Tell me about LangChain")

print("Memory Output 1:")
print(memory_output_1)

print("\nMemory Output 2:")
print(memory_output_2)
```

```python
# Step 16: Create a simple tool
# A tool is just a Python function that performs a task

def word_counter_tool(text):
    """
    This tool counts the number of words in a sentence.
    """
    return f"The sentence contains {len(text.split())} words."

# Test the tool
tool_result = word_counter_tool("LangChain helps build modular AI applications")

print("Tool Output:")
print(tool_result)
```

```python
# Step 17: Simulate a very simple agent
# This agent decides whether to use the tool or the LLM

def simple_agent(user_input):
    """
    If the input asks to count words, use the tool.
    Otherwise, use the language model.
    """

    if "count words" in user_input.lower():
        # Remove the instruction part and send remaining text to the tool
        cleaned_text = user_input.replace("count words", "").strip()
        return word_counter_tool(cleaned_text)
    else:
        return llm.invoke(user_input)

# Test agent
agent_result_1 = simple_agent("count words LangChain makes AI development easier")
agent_result_2 = simple_agent("Explain what LangChain is")

print("Agent Output 1:")
print(agent_result_1)

print("\nAgent Output 2:")
print(agent_result_2)
```

```python
# Step 18: Simulate document loading
# In a real system, LangChain can load PDFs, text files, or web pages.
# Here we simulate it using plain text.

document_text = """
LangChain is a framework for building applications powered by language models.
It supports prompt templates, chains, memory, tools, and agents.
"""

print("Loaded Document Content:")
print(document_text)
```

```python
# Step 19: Simulate a simple retrieval system
# This is a basic keyword-based search, not a full vector store

def retrieve_information(query, document):
    """
    This function checks whether a query word exists in the document.
    """
    if query.lower() in document.lower():
        return f"Relevant information found for query: '{query}'"
    else:
        return f"No direct information found for query: '{query}'"

retrieval_result = retrieve_information("memory", document_text)

print("Retrieval Output:")
print(retrieval_result)
```

```python
# Step 20: Full pipeline demo
# User Input -> Prompt -> LLM -> Tool/Agent -> Output

user_topic = "LangChain"
user_audience = "beginner"
user_tone = "formal"
user_style = "teaching"

# Generate prompt
pipeline_prompt = generate_prompt(
    topic=user_topic,
    audience=user_audience,
    tone=user_tone,
    style=user_style
)

# Send prompt to model
pipeline_llm_output = llm.invoke(pipeline_prompt)

# Final result
print("Pipeline Prompt:")
print(pipeline_prompt)

print("\nPipeline Output:")
print(pipeline_llm_output)
```
