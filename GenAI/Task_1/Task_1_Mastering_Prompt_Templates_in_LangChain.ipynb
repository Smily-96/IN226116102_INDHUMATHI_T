## Task 1: Replace Hardcoded Prompt

### Problem
We need to convert a fixed prompt into a reusable LangChain `PromptTemplate`.

### Required format
Explain {topic} in simple terms for beginners

### Goal
- Do not use f-strings
- Use `PromptTemplate`
- Make it reusable

```python
# -----------------------------
# TASK 1: Replace hardcoded prompt
# -----------------------------

# Create a reusable prompt template
# The variable {topic} will be filled dynamically
task1_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

# Test the template with one sample topic
task1_output = task1_template.format(topic="Artificial Intelligence")

# Print the result
print("Task 1 Output:")
print(task1_output)
```

## Task 2: Multi-Input Prompt System

### Problem
Create a prompt template that accepts 3 inputs:
- topic
- audience
- tone

### Required format
Explain {topic} for {audience} in a {tone} tone

### Test cases
- AI | beginners | friendly
- Python | kids | fun

```python
# -----------------------------
# TASK 2: Multi-Input Prompt System
# -----------------------------

# Create a prompt template with 3 input variables
task2_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

# Test case 1
output1 = task2_template.format(
    topic="AI",
    audience="beginners",
    tone="friendly"
)

# Test case 2
output2 = task2_template.format(
    topic="Python",
    audience="kids",
    tone="fun"
)

# Test case 3
output3 = task2_template.format(
    topic="Deep Learning",
    audience="engineers",
    tone="technical"
)

# Print all outputs
print("Task 2 Output 1:")
print(output1)

print("\nTask 2 Output 2:")
print(output2)

print("\nTask 2 Output 3:")
print(output3)
```

```python
# -----------------------------
# TASK 3: Prompt Variations Engine
# -----------------------------

# Teaching style template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

# Interview style template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}."
)

# Storytelling style template
storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story."
)

# Use the same topic for all templates
topic_value = "Machine Learning"

# Format outputs
teaching_output = teaching_template.format(topic=topic_value)
interview_output = interview_template.format(topic=topic_value)
storytelling_output = storytelling_template.format(topic=topic_value)

# Print outputs
print("Task 3 - Teaching Prompt:")
print(teaching_output)

print("\nTask 3 - Interview Prompt:")
print(interview_output)

print("\nTask 3 - Storytelling Prompt:")
print(storytelling_output)
```

## Task 4: ChatPromptTemplate System

### Problem
Build a chat-based prompt system using roles:
- teacher
- interviewer
- motivator

### Goal
Use:
- system message
- user message

This task shows how to build role-based prompts in LangChain.

```python
# -----------------------------
# TASK 4: ChatPromptTemplate System
# -----------------------------

def build_chat_prompt(role, topic):
    """
    This function creates a chat prompt based on the selected role and topic.
    """

    # Define the system behavior based on the role
    if role == "teacher":
        system_message = "You are a helpful teacher who explains topics clearly to students."
    elif role == "interviewer":
        system_message = "You are a professional interviewer who asks clear and thoughtful questions."
    elif role == "motivator":
        system_message = "You are a motivational coach who inspires and encourages learners."
    else:
        system_message = "You are a helpful assistant."

    # Create a chat prompt template
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Please help me with the topic: {topic}.")
    ])

    # Format the messages using the topic input
    formatted_messages = chat_template.format_messages(topic=topic)

    return formatted_messages


# Example 1: Teacher role
teacher_output = build_chat_prompt("teacher", "Neural Networks")

print("Task 4 Output - Teacher Role:")
for message in teacher_output:
    print(f"{message.type.upper()}: {message.content}")

print("\n" + "-"*50)

# Example 2: Interviewer role
interviewer_output = build_chat_prompt("interviewer", "Neural Networks")

print("Task 4 Output - Interviewer Role:")
for message in interviewer_output:
    print(f"{message.type.upper()}: {message.content}")

print("\n" + "-"*50)

# Example 3: Motivator role
motivator_output = build_chat_prompt("motivator", "Neural Networks")

print("Task 4 Output - Motivator Role:")
for message in motivator_output:
    print(f"{message.type.upper()}: {message.content}")
```

## Task 5: Input Validation Layer

### Problem
Create a function:
`validate_inputs(audience, tone)`

### Rules
- audience must be one of: beginner, intermediate, expert
- tone must be one of: formal, casual, fun

### If invalid
- either raise an error
- or assign default values

Here, we use default values because it is simple and beginner-friendly.

```python
# -----------------------------
# TASK 5: Input Validation Layer
# -----------------------------

def validate_inputs(audience, tone):
    """
    This function validates audience and tone.
    If an invalid value is given, it assigns a default value.
    """

    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Check audience
    if audience not in valid_audiences:
        print(f"Invalid audience '{audience}' found. Default value 'beginner' is used.")
        audience = "beginner"

    # Check tone
    if tone not in valid_tones:
        print(f"Invalid tone '{tone}' found. Default value 'formal' is used.")
        tone = "formal"

    return audience, tone


# Example 1: Valid input
a1, t1 = validate_inputs("beginner", "fun")
print("Validated Output 1:", a1, t1)

# Example 2: Invalid input
a2, t2 = validate_inputs("kids", "friendly")
print("Validated Output 2:", a2, t2)
```

## Task 6: Prompt Generator App

### Problem
Create a function:
`generate_prompt(topic, audience, tone, style)`

### Requirements
- use `PromptTemplate`
- support styles:
  - teaching
  - interview
  - storytelling

### Example output
Explain Neural Networks for beginners in a fun storytelling style

```python
# -----------------------------
# TASK 6: Prompt Generator App
# -----------------------------

def generate_prompt(topic, audience, tone, style):
    """
    This function validates the inputs and generates a structured prompt
    using PromptTemplate.
    """

    # Validate audience and tone first
    audience, tone = validate_inputs(audience, tone)

    # List of valid styles
    valid_styles = ["teaching", "interview", "storytelling"]

    # Validate style
    if style not in valid_styles:
        print(f"Invalid style '{style}' found. Default value 'teaching' is used.")
        style = "teaching"

    # Convert audience words into more natural output words
    if audience == "beginner":
        final_audience = "beginners"
    elif audience == "intermediate":
        final_audience = "intermediate learners"
    else:
        final_audience = "experts"

    # Create a reusable prompt template
    generator_template = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style."
    )

    # Format and return the prompt
    final_prompt = generator_template.format(
        topic=topic,
        audience=final_audience,
        tone=tone,
        style=style
    )

    return final_prompt


# Example test
task6_output = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print("Task 6 Output:")
print(task6_output)
```

## Task 7: Template Reusability Test

### Problem
Use one template and run it with 5 different inputs.

### Goal
- same structure
- different outputs

This proves that the prompt template is reusable.

```python
# -----------------------------
# TASK 7: Template Reusability Test
# -----------------------------

# Create one reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

# Define five different test inputs
test_inputs = [
    {"topic": "AI", "audience": "beginners", "tone": "formal"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Data Science", "audience": "students", "tone": "casual"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"},
    {"topic": "Natural Language Processing", "audience": "researchers", "tone": "formal"}
]

# Run the same template with 5 different inputs
print("Task 7 Outputs:")
for i, item in enumerate(test_inputs, start=1):
    result = reusable_template.format(
        topic=item["topic"],
        audience=item["audience"],
        tone=item["tone"]
    )
    print(f"\nOutput {i}:")
    print(result)
```

## Final Pipeline Demo

### Expected flow
User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output

This final demo combines the full process into one example.

```python
# -----------------------------
# FINAL PIPELINE DEMO
# -----------------------------

# Step 1: User input
user_topic = "Large Language Models"
user_audience = "beginner"
user_tone = "fun"
user_style = "teaching"

# Step 2: Generate the final prompt
final_result = generate_prompt(
    topic=user_topic,
    audience=user_audience,
    tone=user_tone,
    style=user_style
)

# Step 3: Display the output
print("Pipeline Output:")
print(final_result)
```
